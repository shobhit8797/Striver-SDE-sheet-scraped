```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://takeuforward.org/interviews/strivers-sde-sheet-top-coding-interview-problems/"
```

```python
html_doc = requests.get(url).text
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
dict = {}
```

```python
def get_questions(soup):
    dict = {}

    for i in soup.find_all('details'):
        # i.find("tbody").find_all("tr")
        topic = i.find('summary').text.split(":")[1].strip()
        dict[topic] = {}
        a = 1
        if i.find("tbody") is None:
            continue
        for j in i.find("tbody").find_all("tr")[1:]:
            links = j.find_all('td')
            dict[topic][a] = [links[1].text]
            
            for k in links[1:]:
                if k.find('a') is not None:
                    dict[topic][a].append(k.find('a')['href'])
            
            a += 1
        
    return dict
```

```python
data_set = get_questions(soup)
```

```python
df = pd.DataFrame(data=data_set)
```

```python
new_data = {}
for i in data_set:
    for j in data_set[i]:
        for k in range(len(data_set[i][j])):
            if k == 0:
                new_data[data_set[i][j][k]] = {"topic": i}                
            else:
                new_data[data_set[i][j][0]]["link{}".format(k)] = data_set[i][j][k]
new_data
```

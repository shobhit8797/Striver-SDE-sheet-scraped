```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://takeuforward.org/strivers-a2z-dsa-course/strivers-a2z-dsa-course-sheet-2/"
```

```python
html_doc = requests.get(url).text
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
li = []
```

```python
topic = ""
for idx, i in enumerate(soup.find_all('details')):
    dict = {}
    dict['heading'] = i.summary.text
    if i.has_attr('closed'):
        dict['topic'] = topic
        for jidx, j in enumerate(i.find_all('tr')[1:]):
            for kidx, k in enumerate(j.find_all('a')):
                dict[kidx] = k['href']

        li.append(dict)
    else:
        topic = i.summary.text
```

```python
# convert into dataframe

df = pd.DataFrame(data=li)
```
